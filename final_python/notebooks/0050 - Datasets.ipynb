### Sonar (input numerical, output categorical - binary)
The Sonar Dataset involves the prediction of whether or not an object is a mine or a rock given the strength of sonar returns at different angles. It is a binary (2-class) classification problem. The number of observations for each class is not balanced. There are 208 observations with 60 input variables and 1 output variable. The variable names are as follows:
- Sonar returns at different angles (columns 0-59)
- Class (M for mine and R for rock) - column 60

````python
df = pd.read_csv('../data/sonar.csv', header=None)
df.head()
````

````python
#y = df[[60]]
#X = df.drop(columns=60)
df[60]
````

````python
cols = df.columns.to_list()
cols[55:]
````

````python
# Column names should start with a letter, otherwise some libraries throw errors
cols[60] = f"{'class'}" # target (dependent variable)
cols[0:60] = add_prefix_to_list_items('v', cols[:60]) # independent variables
cols[55:]
````

````python
df.columns=cols
df.sample(5)
````

````python
add_dataset(data=df, name="sonar", url="../data/", type="csv", origin="The Sonar Dataset involves the prediction of whether or not an object is a mine or a rock given the strength of sonar returns at different angles: https://archive.ics.uci.edu/ml/datasets/Connectionist+Bench+(Sonar,+Mines+vs.+Rocks)")
````

### Breast Cancer Categorical Dataset (input categorical, output categorical - binary)

The [Breast Cancer Categorical Dataset](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer) dataset classifies breast cancer patient data as either a recurrence or no recurrence of cancer. There are 286 examples and nine input variables. All nine input variables are categorical. It is a **binary classification problem**. A naive model can achieve an accuracy of 70% on this dataset. A good score is about 76% +/- 3%.

````python
cols = ['age', 'menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'deg-malig', 'breast', 'breast-quad', 'irradiat', 'class']
df = pd.read_csv('../data/breast-cancer.data', names=cols)
df.head()
````

````python
df.dtypes
````

````python
# Remove '' from all columns
for col in cols:
    df[col] = df[col].map(lambda x: str(x).rstrip("'").lstrip("'"))
````

````python
df.to_csv("../data/breast-cancer.csv", index=False)
````

````python
add_dataset(data=df, name="breast-cancer", url="../data/", type="csv", origin="Breast cancer patient data as either a recurrence or no recurrence of cancer")
````

````python
# target = 'class'
df = get_dataset('breast-cancer')
df.head()
````
